Ambas as séries de preços apresentam Valor de P alto, refutando a hipótese de estacionariedade.

<h2><center>Previsão de valores com ARIMA</center></h2>
<hr>

<hr>
<p><center>Etanol</center></p>
<hr>

```python
# Configurações
n_periodos_previsao = 146  # Previsão para 2019 em diante
```

```python
arima_etanol = auto_arima(df_etanol_resumo, start_p=1, start_q=1, max_p=6, max_q=6, 
                          m=52, d=1, seasonal=True, trace=True, error_action='ignore', suppress_warnings=True, stepwise=True)
```

```python
treino = df_etanol_resumo.iloc[:-n_periodos_previsao]
teste = df_etanol_resumo.iloc[-n_periodos_previsao:]
arima_etanol.fit(treino)

print(arima_etanol.summary())
```

```python
previsao = arima_etanol.predict(n_periods=n_periodos_previsao)
previsao = pd.DataFrame(previsao,index = teste.index, columns=['preco_venda'])
int_conf = arima_etanol.predict(len(teste.preco_venda), return_conf_int=True, alpha = 0.05)
previsao
```

```python
plt.figure(figsize=(14,8))
plt.title('Previsão para o preço do etanol utilizando ARIMA')
plt.plot(treino, color='green', label = 'Dados de treinamento')
plt.plot(teste, color = 'blue', label = 'Dados de teste')
plt.fill_between(teste.index, int_conf[1][:,0], int_conf[1][:,1], color='k', alpha=.08)
plt.plot(previsao, color = 'red', label = 'Dados previstos')
plt.xlabel('Período')
plt.ylabel('Reais')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.title('Previsão para o preço do etanol utilizando ARIMA')
plt.plot(teste, color = 'blue', label = 'Dados de teste')
plt.fill_between(teste.index, int_conf[1][:,0], int_conf[1][:,1], color='k', alpha=.08)
plt.plot(previsao, color = 'red', label = 'Dados previstos')
plt.xlabel('Período')
plt.ylabel('Reais')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Cálculo do erro
mse = mean_squared_error(teste, previsao)
print('MSE: '+str(mse))
mae = mean_absolute_error(teste, previsao)
print('MAE: '+str(mae))
rmse = math.sqrt(mean_squared_error(teste, previsao))
print('RMSE: '+str(rmse))
```

```python
residuos = pd.DataFrame(arima_etanol.resid())
plt.plot(residuos, color = 'red')
plt.xlabel('Período')
plt.ylabel('Resíduos')
plt.title('Resíduos da previsão')
plt.show()
```

```python
# Gráfico de distribuição dos resíduos
residuos.plot(kind='kde', legend=None, color = 'red')
plt.title('Densidade dos resíduos')
plt.show()
print(residuos.describe())
```

```python
# Correlogramas dos resíduos
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8,8))
plot_acf(residuos, ax = ax1)
plot_pacf(residuos, ax = ax2)
plt.show()
```

```python
# Criação da estrutura de dados com janelas
X_treino = []
y_treino = []

for i in range(janela, len(etanol_treino)):
    X_treino.append(etanol_treino[i-janela:i, :])
    y_treino.append(etanol_treino[i, :])
    
X_treino, y_treino = np.array(X_treino), np.array(y_treino)
```

```python
print("Formato de X_treino: ", X_treino.shape)
print("Formato de y_treino: ", y_treino.shape)
```

```python
# Modelo LSTM - univariado

Model_uni = Sequential()

Model_uni.add(LSTM(units = 60, return_sequences = True, input_shape = (X_treino.shape[1],1)))
Model_uni.add(Dropout(0.2))

Model_uni.add(LSTM(units = 60, return_sequences = True))
Model_uni.add(Dropout(0.2))

Model_uni.add(LSTM(units = 60))
Model_uni.add(Dropout(0.2))

Model_uni.add(Dense(units = 1))

Model_uni.compile(optimizer = 'adam', loss = 'mean_squared_error')

Model_uni.fit(X_treino, y_treino, epochs = 150, batch_size = 16)

Model_uni.summary()
```

```python
plot_model(Model_uni, show_shapes=True)
```

```python
plt.plot(range(len(Model_uni.history.history['loss'])),Model_uni.history.history['loss'] )
plt.xlabel('Epoch Number')
plt.ylabel('Loss')
plt.show()
```

```python
# Previsões
previsao = []
lote_um = etanol_treino[-janela:]
novo_lote = lote_um.reshape((1,janela,1))

for i in range(n_periodos_previsao):
    previsao_unitaria = Model_uni.predict(novo_lote)[0]
    previsao.append(previsao_unitaria)
    novo_lote = np.append(novo_lote[:,1:,:], [[previsao_unitaria]], axis = 1)
    
previsao = np.array(previsao)
previsao = scaler.inverse_transform(previsao)
```

```python
# Desnormalizando valores de treino e teste
etanol_treino = scaler.inverse_transform(etanol_treino)
etanol_teste = scaler.inverse_transform(etanol_teste)
```

```python
# Visualizando previsões e dados observados

teste_plot = pd.DataFrame(etanol_teste[:,0], df_etanol_resumo.iloc[-n_periodos_previsao:,:].index)
previsao_plot = pd.DataFrame(previsao, df_etanol_resumo.iloc[-n_periodos_previsao:,:].index)
treino_plot = pd.DataFrame(etanol_treino[:,0], df_etanol_resumo.iloc[:-n_periodos_previsao,:].index)

plt.plot(teste_plot, color='green', label = 'Dados de treinamento')
plt.plot(treino_plot, color = 'blue', label = 'Dados de teste')
plt.plot(previsao_plot, color = 'red', label = 'Valores previstos')
plt.title('Previsão para o preço do etanol utilizando RN LSTM')
plt.xlabel('Período')
plt.ylabel('Reais')
plt.legend()
plt.show()
```

```python
# Cálculo do erro
print('MAE: ', mean_absolute_error(etanol_teste, previsao))
print('MSE: ', mean_squared_error(etanol_teste,previsao))
print('RMSE: ', np.sqrt(mean_squared_error(etanol_teste,previsao)))
```

```python
# Normalização
scaler = MinMaxScaler(feature_range = (0,1))
etanol_multi_scaled = scaler.fit_transform(etanol_multi)
```

```python
# Divisão do conjunto de dados em treino e teste
treino_multi = etanol_multi_scaled[:-n_periodos_previsao,:]
teste_multi = etanol_multi_scaled[-n_periodos_previsao:,1]
```

```python
# Criação da estrutura de dados com janelas
X_treino_m = []
y_treino_m = []

for i in range(janela, len(treino_multi)):
    X_treino_m.append(treino_multi[i-janela:i, 0:2])
    y_treino_m.append(treino_multi[i, 1])
    
X_treino_m, y_treino_m = np.array(X_treino_m), np.array(y_treino_m)
```

```python
# Modelo LSTM - multivariado

Model_multi = Sequential()

Model_multi.add(LSTM(units = 60, return_sequences = True, input_shape = (X_treino_m.shape[1], 2)))
Model_multi.add(Dropout(0.2))

Model_multi.add(LSTM(units = 60, return_sequences = True))
Model_multi.add(Dropout(0.2))

Model_multi.add(LSTM(units = 60))
Model_multi.add(Dropout(0.2))

Model_multi.add(Dense(units = 1))

Model_multi.compile(optimizer = 'adam', loss = 'mean_squared_error')

Model_multi.fit(X_treino_m, y_treino_m, epochs = 150, batch_size = 16)

Model_multi.summary()
```
