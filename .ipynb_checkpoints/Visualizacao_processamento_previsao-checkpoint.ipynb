<h2>Prevendo valores com AUTO-ARIMA</h2>

<p>A previsão com o modelo ARIMA é realizada para a série de homicídios</p>

```python
# Correlogramas: ACF e PACF
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8,8))
plot_acf(summr_homicidio, ax = ax1)
plot_pacf(summr_homicidio, ax = ax2)
plt.show()
```

A autocorrelação (ACF) decai lentamente, enquanto a autocorrelação parcial tem picos em h=1 e em h=7,8,13,14, 16.17,19,21, o que sugere a investigação de um modelo ARIMA com sazonalidade e p=1.

```python
#Análise de estacionariedade para summr_homicidio
X = summr_homicidio['Registros']
result = adfuller(X)
print('ADF Estatíticas: %f' % result[0])
print('Valor de P: %f' % result[1])
print('Valores Críticos:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

A série de homicídios apresenta Valor de P alto, refutando a hipótese de estacionariedade. 

```python
# Diferenciando a série
homicidio_diff = summr_homicidio - summr_homicidio.shift()
homicidio_diff.dropna(inplace=True)

# Gráfico da série diferenciada
plt.figure(figsize=(16,8))
plt.plot(homicidio_diff)
plt.title('Série de homicídios diferenciada')
plt.grid(True)
```

Aparentemente, a série se tornou estacionária com a diferenciação de primeira ordem. 

```python
#Análise de estacionariedade para summr_homicidio
X = homicidio_diff['Registros']
result = adfuller(X)
print('ADF Estatíticas: %f' % result[0])
print('Valor de P: %f' % result[1])
print('Valores Críticos:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

O teste de ADFuller, com um Valor de P baixo, aceita a hipótese de estacionariedade após diferenciação, o que sugere um modelo ARIMA com d=1.

<h2>Prevendo valores com Rede Neural</h2>

```python
# Configurações
n_periodos_previsao = 12
window = 12
```

```python
# Normalização
scaler = MinMaxScaler(feature_range = (0,1))
homicidio_scaled = scaler.fit_transform(summr_homicidio)
```

```python
# Divisão do conjunto de dados em treino e teste
training_set = homicidio_scaled[:-n_periodos_previsao,:]
test_set = homicidio_scaled[-n_periodos_previsao:,:]
```

```python
# Criação da estrutura de dados com janelas
X_train = []
y_train = []

for i in range(window, len(training_set)):
    X_train.append(training_set[i-window:i, 0:1])
    y_train.append(training_set[i, 0])
    
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
print("Formato de X_train: ", X_train.shape)
print("Formato de y_train: ", y_train.shape)
```

```python
# Modelo LSTM - univariado

Model_uni = Sequential()

Model_uni.add(LSTM(units = 60, return_sequences = True, input_shape = (X_train.shape[1],1)))
Model_uni.add(Dropout(0.2))

#Model_uni.add(LSTM(units = 60, return_sequences = True))
#Model_uni.add(Dropout(0.2))

#Model_uni.add(LSTM(units = 60, return_sequences = True))
#Model_uni.add(Dropout(0.2))

Model_uni.add(LSTM(units = 60))
Model_uni.add(Dropout(0.2))

Model_uni.add(Dense(units = 1))

Model_uni.compile(optimizer = 'adam', loss = 'mean_squared_error')

Model_uni.fit(X_train, y_train, epochs = 150, batch_size = 16)

Model_uni.summary()
```

```python
plt.plot(range(len(Model_uni.history.history['loss'])),Model_uni.history.history['loss'] )
plt.xlabel('Epoch Number')
plt.ylabel('Loss')
plt.show()
```

```python
# Previsões
prediction_test = []
batch_one = training_set[-window:]
batch_new = batch_one.reshape((1,window,1))

for i in range(n_periodos_previsao):
    first_pred = Model_uni.predict(batch_new)[0]
    prediction_test.append(first_pred)
    batch_new = np.append(batch_new[:,1:,:], [[first_pred]], axis = 1)
    
prediction_test = np.array(prediction_test)
predictions = scaler.inverse_transform(prediction_test)
```

```python
# Desnormalizando valores de treino e teste
training_set = scaler.inverse_transform(training_set)
test_set = scaler.inverse_transform(test_set)
```

```python
# Visualizando previsões e dados observados
test_set = pd.DataFrame(test_set[:,0], summr_homicidio.iloc[-n_periodos_previsao:, :].index)
predictions = pd.DataFrame(predictions[:,0], summr_homicidio.iloc[-n_periodos_previsao:, :].index)

plt.plot(test_set, color = 'green', label = 'Valores observados')
plt.plot(predictions, color = 'blue', label = 'Valores previstos')
plt.title('LSTM - Univariate Forecast')
plt.xlabel('Meses')
plt.ylabel('Número de ocorrências')
plt.legend()
plt.show()
```

```python
plt.plot(summr_homicidio, label='Observado', color='green');
plt.plot(predictions, label='Previsão para os Dados', color='blue');
plt.legend(loc='best');
plt.title('Rede Neural Artificial')
plt.show()
```

```python
# Cálculo do erro e R²
print('MAE: ', mean_absolute_error(test_set,predictions))
print('MSE: ', mean_squared_error(test_set,predictions))
print('RMSE: ', np.sqrt(mean_squared_error(test_set,predictions)))
```

<h2>LSTM Multivariada</h2>

```python
# Configurações
n_periodos_previsao = 12
window = 12
```

<h4>Homicídios e valor investido em forças de segurança</h4>

```python
# Seleção dos dados de homicídios
dfhomicidio = dfconjunto.drop(['Roubo', 'Extorsao', 'Sequestro', 'Estupro'], axis=1)
dfhomicidio_scaled = dfconjunto_scaled.drop(['Roubo', 'Extorsao', 'Sequestro', 'Estupro'], axis=1)
```

```python
# Divisão do conjunto de dados em treino e teste
h_training_set = dfhomicidio_scaled.iloc[:-n_periodos_previsao,:].values
h_test_set = dfhomicidio_scaled.iloc[-n_periodos_previsao:,0].values
```

```python
# Criação da estrutura de dados com janelas
X_train_h = []
y_train_h = []

for i in range(window, len(h_training_set)):
    X_train_h.append(h_training_set[i-window:i, 0:2])
    y_train_h.append(h_training_set[i, 1])
    
X_train_h, y_train_h = np.array(X_train_h), np.array(y_train_h)
```

```python
print("Formato de X_train_h: ", X_train_h.shape)
print("Formato de y_train_h: ", y_train_h.shape)
```

```python
# Modelo LSTM - multivariado

Model_multi = Sequential()

Model_multi.add(LSTM(units = 60, input_shape = (X_train_h.shape[1], 2)))
Model_multi.add(Dropout(0.2))

#Model_multi.add(LSTM(units = 60))
#Model_multi.add(Dropout(0.2))

Model_multi.add(Dense(units = 1))

Model_multi.compile(optimizer = 'adam', loss = 'mean_squared_error')

Model_multi.fit(X_train_h, y_train_h, epochs = 500, batch_size = 1)

Model_multi.summary()
```

```python
plt.plot(range(len(Model_multi.history.history['loss'])), Model_multi.history.history['loss'])
plt.xlabel('Epoch Number')
plt.ylabel('Loss')
plt.show()
```

```python
# Prevendo os valores
prediction_test = []

batch_one = h_training_set[-window:]
batch_new = batch_one.reshape((1,window,2))

for i in range(n_periodos_previsao):
    first_pred = Model_multi.predict(batch_new)[0]
    prediction_test.append(first_pred)
    aux = h_test_set[i]
    aux = aux.reshape(1,1)
    new_test = np.insert(aux, 1, first_pred, axis = 1)
    new_test= new_test.reshape(1,1,2)
    batch_new = np.append(batch_new[:, 1:,:], new_test, axis=1)
    
prediction_test = np.array(prediction_test)
```

```python
sc = MinMaxScaler(feature_range = (0,1))
y_scale = dfhomicidio.iloc[:,1:2]
sc.fit_transform(y_scale);
```

```python
predictions = sc.inverse_transform(prediction_test)
predictions = pd.DataFrame(predictions, dfhomicidio.iloc[-n_periodos_previsao:,1].index)
h_test_set = pd.DataFrame(dfhomicidio.iloc[-n_periodos_previsao:,1])
```

```python
plt.plot(h_test_set, color = 'red', label = 'Valores Observados')
plt.plot(predictions, color = 'blue', label = 'Valores Previstos')
plt.title('Número de homicídios')
plt.xlabel('Mês')
plt.ylabel('Número')
plt.legend()
plt.show()
```

```python
plt.plot(dfhomicidio.iloc[:,1], label='Observado', color='green');
plt.plot(predictions, label='Previsão para os Dados', color='blue');
plt.legend(loc='best');
plt.title('Rede Neural Artificial')
plt.show()
```

```python
# Cálculo do erro e R²
print('MAE: ', mean_absolute_error(h_test_set,predictions))
print('MSE: ', mean_squared_error(h_test_set,predictions))
print('RMSE: ', np.sqrt(mean_squared_error(h_test_set,predictions)))
print('R2: ', r2_score(h_test_set, predictions))
```
